### Filling Missing Values

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data['Life expectancy '].fillna(data['Life expectancy '].mean(), inplace=True)

data['Adult Mortality'].fillna(data['Adult Mortality'].mean(), inplace=True)

data['Alcohol'].fillna(data['Alcohol'].mean(), inplace=True)

data['Hepatitis B'].fillna(data['Hepatitis B'].mean(), inplace=True)

data[' BMI '].fillna(data[' BMI '].mean(), inplace=True)

data['Polio'].fillna(data['Polio'].mean(), inplace=True)

data['Total expenditure'].fillna(data['Total expenditure'].mean(), inplace=True)
```

```python
data['Diphtheria '].fillna(data['Diphtheria '].mean(), inplace=True)

data['GDP'].fillna(data['GDP'].mean(), inplace=True)

data['Population'].fillna(data['Population'].mean(), inplace=True)

data[' thinness  1-19 years'].fillna(data[' thinness  1-19 years'].mean(), inplace=True)

data[' thinness 5-9 years'].fillna(data[' thinness 5-9 years'].mean(), inplace=True)

data['Income composition of resources'].fillna(data['Income composition of resources'].mean(), inplace=True)

data['Schooling'].fillna(data['Schooling'].mean(), inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
data['Country'].nunique()
```

```python
data['Country'].value_counts()
```

```python
data['Status'].value_counts()
```

```python

data['Status'] = data['Status'].replace({'Developing' : '0', 'Developed' : '1'}) 
```

```python
data.sample(10)
```

```python
# one hot encoding for column 'Country'

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

data['Country'] = ohe.fit_transform(data['Country'].values.reshape(-1, 1)).toarray()
```

```python
data.sample(5)
```

### Train Test Split

```python
X = data.drop('Life expectancy ', axis=1)
y = data['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=15)
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

sd = StandardScaler()

Xtrain_scaled = sd.fit_transform(Xtrain)
Xtest_scaled = sd.fit_transform(Xtest)
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(Xtrain, ytrain)
```

```python
ypred1 = lr.predict(Xtest)
```

```python
ypred1
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

r2_score(ytest, ypred1)
```

```python
mean_squared_error(ytest, ypred1)
```

```python
mean_absolute_error(ytest, ypred1)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=10)

dt.fit(Xtrain, ytrain)
```

```python
ypred2 = dt.predict(Xtest)
```

```python
ypred2
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

r2_score(ytest, ypred2)
```
